<p dir = 'rtl'>  برای خوانایی بیشتر ما میتونیم دستور بالا را در یک متغییر هم قرار بدیم </p>

<p dir = 'rtl'>  مانند دستور های زیر </p>

```python
high_score_movies = (df2['score'] >= 8.8)
df2[high_score_movies]
```

<p dir = 'rtl'>  در دستور زیر ما میایم دوتا متغییر تعریف میکنیم و میگیم فیلم هایی که نمرشون بین 7 و 8 هستند را نمایش بده </p>

```python
cond1 = df2['score'] >= 7
cond2 = df2['score'] <= 8
df2[cond1 & cond2]
```

<p dir = 'rtl'>  همانطور که میدانید با دستور shape نیز میتونیم تعداد ردیف و ستون هارا نشان دهیم </p>

```python
df2[cond1 & cond2].shape
```

<p dir = 'rtl'>  در دستور زیر هم ما دوتا متغییر تعریف میکنیم و فیلم هایی که بعد از سال 2015 نمره بالاتر از  8.5 گرفتند را نمایش دهد </p>

```python
high_score_movies = (df2['score'] >= 8.5)
after_2015 = (df2['year'] >= 2015)
df2[high_score_movies & after_2015]
```
