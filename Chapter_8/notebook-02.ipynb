**ترکیب شرط های منطقی **

```python
import pandas as pd
```

```python
df = pd.read_csv(
    '../Data/nba.csv',
    index_col='Name',
    parse_dates=['Birthday'],
)
df.head()
```

<p dir=rtl>ما قبلا شرط -پی اف - رو نوشتیم بعد میتونیم یه شرط دیگه هم بنویسیم که فرضا تاریخ تولد شون رو داشته باشیم </p>

```python
is_pf = (df['Position'] == 'PF')
after_1998 = (df['Birthday'] >= '1998-01-01') #افرادی که متولد این سال به بعد هستند 
```

```python
after_1998
```

<p dir=rtl>بخش هایی رو انتخاب کن که هم -پی اف - هستند --و-- هم اینکه تاریخ تولدشون این هست</p>

```python
df[is_pf & after_1998] 
```

```python
df[~is_pf] #انهایی که پی اف نیستن (نقیض)
```

```python
df[is_pf | after_1998] #یا پی اف باشن یا تاربخ تولدشون اون باشه 
```

<p dir = rtl>پس ما میتوانیم انواع ترکیب هارو نه رو یک شرط روی چندتا شرط میتوانیم ترکیب کنیم و بهترین حالت هم این هست که ما شرط هامون رو بنویسیم و در یک سری متغیر ذخیره کنیم و بعد آنهارو موقع ایندکس کردن اینجا بفرستیم</p>
