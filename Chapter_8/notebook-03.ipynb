**متد های کاربردی برای تعریف شرط ها**

```python
import pandas as pd
```

<p dir = rtl>ما یاد گرفتیم که چطور میتونیم شرط های مختلف رو چجوری باهم ترکیب کنیم که فیلتر های پیچیده تر بنویسیم و روش کلی این استفاده از تعریف شرط و استفاده از اپراتور های منطقی است برای نقیض ترکیب عطفی و ترکیب فصلی و برای تعریف بعضی از اینها یک سری راهکارهای ساده تر هم وجود دارد </p>

```python
df = pd.read_csv(
    '../Data/movies.csv',
    index_col='name',
    parse_dates=['release_date'],
)
df.head()
```

```python
above_7 = (df['score'] >= 7) # انتخاب امتیاز های بالای  هفت و کمتر از هشت 
below_8 = (df['score'] <= 8)
df[above_7 & below_8]
```

<p dir=rtl>برای ساده تر کردن کد که میگه بین هفت و هشت را نشان بده .</p>

```python
df['score'].between(7,8) 
```

```python
df[df['score'].between(7,8)] #نمایش 
```

```python
df[df['release_date'].between('2010-01-01','2016-01-01')] # با استفاده از تاریخ هم امکان پذیر است  
```

```python
df['genre'] #نمایش ژانر
```

```python
df['genre'].unique() #مقادیر یونیک ژانر رو نمایش میده 
```

```python
is_mystery = (df['genre'] == 'Mystery') # معمایی
is_scifi = (df['genre'] == 'Sci-Fi') # علمی تخیلی
df[is_mystery | is_scifi] # از یا استفاده میکنیم اگر اند بزاریم خالی نمایش میدهد
```

<p dir=rtl>راهکار ساده تر :راهکار ساده تر میگه که چک بکن این ژانر رو اگر عضوی از این مجموعه هست.</p>

```python
df['genre'].isin(['Mystery', 'Sci-Fi']) 
```

<p dir = rtl>در این روش اگر چیزی بخواهیم اضافه کنیم فقط در اول کد اضافه میکنیم</p>

```python
genres = ['Mystery', 'Sci-Fi'] #چیز هایی که لازم داریم
genre_filter = df['genre'].isin(genres) # و شرط
df[genre_filter] 
```

```python
genres = ['Mystery', 'Sci-Fi', 'Adventure'] # اضافه کردن یک ژانر
genre_filter = df['genre'].isin(genres)
df[genre_filter]
```

<p dir=rtl>زمانی هست که ما میخواهیم مقادیری که -نال- هستند رو انتخاب کنیم و یا بر عکس</p>

```python
df['score'].isnull() # مقادیری که نال هستند رو نمایش میدهد 
```

```python
df[df['score'].isnull()] # مواردی که در قسمت امتیاز نال هستند 
```

```python
df['score'].notnull() # برعکس نال یعنی نال نیستند 
```

```python
df[df['score'].notnull()]
```

```python
df[~df['score'].isnull()] #میتوان آن راهم نقیض کرد 
```
