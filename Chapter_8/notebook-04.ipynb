```python
df = pd.read_csv('../Data/used_cars.csv')
df.head()
```

<p dir=rtl>اگر مشاهده کنیم در اینجا قسمت -رنج ایندکس- تعداد -انتری - هارو که کامل هست و همشون مقدار دارند </p>

```python
df.info() 
```

<p dir=rtl>در این قسمت سوخت مشخص نیست و نال هست </p>

```python
df[df['Fuel'].isnull()] 
```

```python
df[df['Transmission'].isnull()] # در اینجا همین گونه است
```

<p dir=rtl> در اینجا از عمگر مقایسه ایی استفاده میکینم و میگیم این یا اون بلاخره یکیشون مشکل داره </p>

```python
df[df['Fuel'].isnull() | df['Transmission'].isnull()]
```

```python
df[df['Fuel'].isnull() & df['Transmission'].isnull()] # در این قسمت از اشتراک انها را با اند میگیریم 
```

```python
df2 = df.fillna('--') #برای جایگزین کردن 
```

```python
df2['Fuel'].value_counts() # مقدار نال ها را با (__) نمایش میدهد
```

```python
df['Fuel'].value_counts() # در اینجا مقدار نال هارو به ما نمایش نمی دهد 
```

```python
fuel = df['Fuel'] 
```

```python
fuel.fillna('N/A') # جایگزین کردن کردن در ستون 
```

```python
df2['Fuel'] = fuel.fillna('N/A') # اینجا میگیم که دی اف 2 فیولش بشه ان ای
```

```python
df2['Fuel'].value_counts() # مشاهده میکنیم که تغییر کرده
```

<p dir = rtl > در قسمت -- هر تغییری که کرده بود در قسمت دیتا فریم ها بود و تغییر داده بود ولی در اینجا هر ستون رو جدا جدا میتونیم- فیل ان ای- کنیم و جاگذاریش کنیم </p>

<p dir = rtl> این روش بعدی ما هست که کلا دروپ بشه بگیم که من نمیخوام هیچ چیزی شبیه این رو که مثلا نال باشه یا هرچیز دیگه ایی این دیتا فریم من میتونه دروپ ان ای بشه که همه مواردی که یک چیز نال داشتن حذف شد که میتونه متد داشته باشه </p>

```python
df.dropna() # در اینجا مشاهده میکنیم همه مواردی که نال داشتند حذف شده 
```

```python
df.dropna(how='all') # وقتی دروپ کن که همه موارد میسینگ باشه (هیچ چیزی رو دروپ نمیکنه )
```

```python
df.dropna(how='any') #در اینجا میگیم حتی اگه یدونه هم نال باشه کافیه 
```

<p dir=rtl>در قسمت -آل- اگر چیزی رو -دروپ- نکرد اینجوری هست که میگه خیلی سخت گیرانه است چون همشون باید -نال- باشند </p>

```python
df.dropna(how='all', subset=['Fuel','Transmission']) #محدود کردن به بعضی از ستون ها 
```

<p dir=rtl>میگه داده ایی که من دارم حداقل این چیز رو باید داشته باشه</p>

```python
df.dropna(thresh=3) # برای بهتر شدن کد از این روش استفاده میکنیم 
```

<p dir = rtl >در تریشولد میگیم حداقل این تعداد رو باید داشته باشه </p>

```python
df.dropna(thresh=4) 
```

<p dir = rtl> این هم میتواند با متود های دیگه ترکیب بشه و -فکوس- رو گذاشت روی چندتا ستون  </p>

```python
df.dropna(thresh=2, subset=['Fuel', 'Transmission'], inplace=True)
```
