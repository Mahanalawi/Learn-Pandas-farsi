**مقادیر تکراری**

```python
import pandas as pd
```

<p dir = rtl> پانداس اماکانت جالبی رو داره برای برخورد با مقادیر تکراری خیلی پیش میاد تو مواردی دیتا ست تکراری دارید حالا به هر نحوی میخواهید موارد تکراری رو حذف کنید یا کلا حذف کنید هیچی از اون مقادیر تکراری نباشه یا اینکه یکیشو نگهدارید برای اینها امکانات خوبی رو داره  </p>

```python
df = pd.read_csv(
    '../Data/employees.csv',
    index_col='First Name',
    parse_dates=['Start Date'],
)
df.head()
```

```python
df['Team'].duplicated() # برای نشان دادن مقادیر تکراری
```

```python
~df['Team'].duplicated() # آنرا نقیض میکنیم
```

```python
df[~df['Team'].duplicated()] #در اینجا از هر تیم یک نفر رو میاره 
```

```python
df.sort_values(by='Start Date', inplace=True) #مرتب سازی بر اساس تاریخ
```

```python
df.head() # از قدیم به جدید میاره 
```

```python
df[~df['Team'].duplicated()] # در اینجا قدیمی ترین نیروی هر تیم رو میاره
```

```python
df[~df['Team'].duplicated(keep='last')] #در اینجا ما میگیم اخرین مورد تکرار رو برای من نگهدار
```

<p dir=rtl>یه زمانی هست مقدار تکراری رو دوست ندارید داشته باشید این رو باید حذف کنید به هر ترتیبی</p>

```python
df.drop_duplicates() # در اینجا هیچ مقادیری و دروپ نمیکند و فقط از هر مقدار یکی و نشون میده 
```

```python
df.drop_duplicates(subset=['Team']) # اگر بهش بگیم جایی که تیم تکرار شده بود اونهارو حذف کن 
```

```python
df.drop_duplicates(subset=['Gender', 'Team']) # از هر تیم یک نفر اقا یا خانم رو مشخص میکند به شرطی حذف میکنه که هر دو انها باهم برابر باشند
```

```python
df.drop_duplicates(keep='last', subset=['Gender', 'Team']) # جدید ترین نیروی هر بخش 
```

<p dir = rtl>   نگه ندار و همرو حذف کن (last و firt)در اینجا یعنی نه (false)  معنی  </p>

```python
df.drop_duplicates(keep=False, subset=['Start Date']) # نفراتی هستند که روزی که اینها امدن کسی با اینها نیومده شرکت 
```

```python
df.drop_duplicates(keep=False, subset=['First Name']) #چون به عنوان ایندکس استفاده کردیم خطا میده 
```

```python
df.reset_index(inplace=True) # از ایندکس بودن خارجش میکنیم 
```

```python
df.drop_duplicates(keep=False, subset=['First Name']) # افرادی هستند که اسم تکراری ندارند
```
