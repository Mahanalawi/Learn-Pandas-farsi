<p dir=rtl> در اینجا میگیم که مقدار در آمد فیلم تقسیم بر بودجه که برای فیلم شده منهای یک  </p>

```python
ROI = movies.gross / movies.budget - 1
```

```python
ROI = (movies.gross - movies.budget) / movies.budget # روش دیگر
```

```python
ROI
```

<p dir=rtl> میتوان خودش رو یک ستون در نظر گرفت و اگر خروجی بگیریم یک ستون اضافه شده است  </p>

```python
movies['ROI'] = movies.gross / movies.budget - 1
```

```python
movies.head()
```

```python
movies.plot(x='release_date', y='ROI') #دوتا فیلم هست حالا با بیشترین بازه وجود داره 
```

```python
movies.sort_values(by='ROI', ascending=False) # مرتب سازی پر بازه ترین فیلم ها 
```

```python
movies.sort_values(by='ROI', ascending=False, inplace=True) 
```

```python
m2 = movies.iloc[50:] # از پنجاه به بعد رو ببینیم 
```

```python
m2.plot(x='release_date', y='ROI') # و ام دو رو پلات میکینم
```
